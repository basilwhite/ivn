# M-26-10 Influence Heatmap

This notebook generates a heatmap of influence for OMB Memorandum M-26-10 on USDA directives. The influence score is calculated based on direct implementation and category relevance.

```python
import os
import sys
import pandas as pd
import plotly.express as px

# Add project root to path
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

os.environ['GRAPH_BACKEND'] = 'neo4j' # or 'kuzu'

from gov_graph.queries import heatmap
```

## Fetch Heatmap Data

```python
heatmap_data = heatmap.get_m2610_influence_heatmap()
df = pd.DataFrame(heatmap_data)
df
```

## Plot the Heatmap

```python
fig = px.bar(df,
             x='directive',
             y='influence_score',
             color='influence_type',
             title='OMB M-26-10 Influence on USDA Directives',
             labels={'directive': 'USDA Directive', 'influence_score': 'Influence Score'},
             category_orders={'influence_type': ['Direct', 'Category']})
fig.show()
```
